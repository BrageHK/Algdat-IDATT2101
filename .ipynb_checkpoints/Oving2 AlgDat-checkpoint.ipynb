```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>
#include <math.h>

time_t start, end;

double power(double x, int n)
{
        if(n == 0) return 0;
        if(n <= 1) return x;
        if(n > 0) return x * power(x, n-1);
        return -1;
}

double power2(double x, int n)
{
    if(n == 0) return 1;
    if((n & 1)==1) return x * power2(x*x,(n-1)/2);   //even
    if((n & 1)==0) return power2(x*x,(n)/2);         //odd
    return x;
}

// Time is in seconds
void timePower(double x, int n, double time)
{
    start = clock(); // Starting clock before running the algorithm
    double num;
    int runder = 0;
    double tid;

    do
    {
        num = power(x,(n));
        end = clock();
        ++runder;
    } while ((double)(end-start)/CLOCKS_PER_SEC < time);
    
    printf("\nUsing algorithm from 2.1-1\nOutput number: %f\n", num);      
    printf("Milliseconds used per round: %f\n", ((((double)(end-start)/CLOCKS_PER_SEC)/runder))*1000);
    printf("Ruonds: %d\n", runder);
}

// Time is in seconds
void timePower2(double x, int n, double time)
{
    start = clock(); // Starting clock before running the algorithm
    double num;
    int runder = 0;
    double tid;

    do
    {
        num = power2(x,(n));
        end = clock();
        ++runder;
    } while ((double)(end-start)/CLOCKS_PER_SEC < time);
    
    printf("\nUsing algorithm from 2.2-3\nOutput number: %f\n", num);      
    printf("Milliseconds used per round: %f\n", ((((double)(end-start)/CLOCKS_PER_SEC)/runder))*1000);
    printf("Ruonds: %d\n", runder);
}

// Time is in seconds
void timePow(double x, int n, double time)
{
    start = clock(); // Starting clock before running the algorithm
    double num;
    int runder = 0;
    double tid;

    do
    {
        num = pow(x,(n));
        end = clock();
        ++runder;
    } while ((double)(end-start)/CLOCKS_PER_SEC < time);
    
    printf("\nUsing Built in pow method:\nOutput number: %f\n", num);      
    printf("Milliseconds used per round: %f\n", ((((double)(end-start)/CLOCKS_PER_SEC)/runder))*1000);
    printf("Ruonds: %d\n", runder);
}


int main(int argc, char *argv[])
{
    // Check if methods work
    if(power(3,14) != 4782969) return -1;
    if(power2(3,14) != 4782969) return -1;
    if(pow(3,14) != 4782969) return -1;
    if(power(2,12) != 4096) return -1;
    if(power2(2,12) != 4096) return -1;
    if(pow(2,12) != 4096) return -1;
    
    // Testing for different values of n
    double x = 1.01;
    int n = 1;
    double time = 2;
    timePower(x, n, time);
    timePower2(x, n, time);  
    timePow(x, n, time);
    
    n = 10;
    timePower(x, n, time);
    timePower2(x, n, time);  
    timePow(x, n, time);

    n = 100;
    timePower(x, n, time);
    timePower2(x, n, time);  
    timePow(x, n, time);

    n = 1000;
    timePower(x, n, time);
    timePower2(x, n, time);  
    timePow(x, n, time);

    n = 10000;
    timePower(x, n, time);
    timePower2(x, n, time);  
    timePow(x, n, time);

    
    
    return 0;
}
```
